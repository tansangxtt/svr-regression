```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("data-15Feb.csv")
dataset
```

```python
X = dataset.iloc[:, 0:6]
X
```

```python
Y = dataset.iloc[:, 6:7]
Y
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_Y = StandardScaler()
X = sc_X.fit_transform(X)
Y = sc_Y.fit_transform(Y)
# X = X.to_numpy()
# Y = Y.to_numpy()
```

```python
X
```

```python
# Y = Y*10**5
Y
```

```python
# spliting the dataset into training set and test set
from sklearn.model_selection import train_test_split
Y = Y.reshape(-1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.svm import SVR

# The most important SVR parameter is the Kernel. We would select RBF (Radial Basis Function) first

C_range = np.logspace(-2, 10, 13)
gamma_range = np.logspace(-9, 3, 13)
regressor_list = []
testing_errors = []
param_list = []

for c_value in C_range:
    for gamma_value in gamma_range:
        regressor = SVR(kernel = 'rbf', gamma=gamma_value, C=c_value)
        regressor.fit(X_train, Y_train)
        regressor_list.append(regressor)

        y_pred = regressor.predict(X)
        y_pred_train = regressor.predict(X_train)
        y_pred_test = regressor.predict(X_test)
        testing_error = np.mean(np.square(np.array(Y_test) - np.array(y_pred_test)))
        testing_errors.append(testing_error)

        param_list.append( [c_value, gamma_value] )


minimum_error_index = 0
for i in range(len(testing_errors)):
    if (testing_errors[i] < testing_errors[minimum_error_index]):
        minimum_error_index = i   


print ("Hello Tan Min, I found the best param here: C {}, gamma {}, error {}"%(param_list[minimum_error_index][0], param_list[minimum_error_index][1]), testing_errors[minimum_error_index])     


# regressor = SVR(kernel = 'rbf', gamma=0.9, C=100)
# regressor.fit(X_train, Y_train)
```
